```python
import pandas as pd
import numpy as np
import math
import random
import webbrowser

## http://download.geonames.org/export/dump/?C=M;O=D
cities = pd.read_csv("/Users/calummeikle/Documents/4th Year(And Beyond)/Test Driven Python/world-cities-master/cities15000.txt", sep='\t',
                     header = None, names =['id','name','name_again', 'random_info', 'latitude_deg', 'longitude_deg', 'P', 'PPL', 'area_code', '1','2','3', '4', '5', 'population', '6', 'random_number', 'area', 'date_added'])
#pd.read_fwf is another option


## https://ourairports.com/data/
airports = pd.read_csv("/Users/calummeikle/Documents/4th Year(And Beyond)/Test Driven Python/world-cities-master/airports.csv")
#Airport data is slightly incomplete

#need to get rid of the heliport and closed airports, use re (didn't need it in the end)

big_cities = cities.sort_values(by=['population'], ascending=False)
big_cities = big_cities.reset_index(drop=True)

#remove all that isn't a passenger airport I may have missed some
airports = airports[airports.type != 'heliport']
airports = airports[airports.type != 'closed']
airports = airports[airports.type != 'seaplane_base']
airports = airports[airports.type != 'air_base']
airports = airports[airports.type != 'airfield']
airports
```

```python
# Pick a random city
x = math.floor(len(big_cities['name'])/20)
big_cities = big_cities[0: x]
y = random.randint(0, x)

city = big_cities.loc[y, 'name']

#finding the closest airport to your random city
position_lat = abs(airports['latitude_deg'] - big_cities.loc[y, 'latitude_deg'])
position_long = abs(airports['longitude_deg'] - big_cities.loc[y, 'longitude_deg'])
min_distance = (position_lat + position_long).argmin()
```

```python
airport = airports.loc[min_distance, 'name']
print("This is your new city break {}.".format(city))
print("And the nearest airport is {}.".format(airport))
```

```python
web_link = "https://en.wikipedia.org/wiki/" + city
webbrowser.open(web_link)
```
